```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cv2
# Exemplo de DataFrame com as classes
df = pd.read_csv("df_ori_mask_crop.csv")
df = df.drop(df.loc[df['Finding Labels'] == 'No Finding'].index) # Removendo No Finding

# Selecionar aleatoriamente 6 imagens e suas classes
sample_df = df.sample(6)
# Plotar as imagens em uma grade 2x3
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
for i, (index, row) in enumerate(sample_df.iterrows()):
    img = plt.imread(row['path'])
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Converter de BGR para RGB
    axs[i // 3, i % 3].imshow(img_rgb)
    axs[i // 3, i % 3].axis('off')
    title = axs[i // 3, i % 3].set_title(row['Finding Labels'], fontsize=13, fontweight='bold')
    title.set_bbox({'facecolor': 'white', 'edgecolor': 'black', 'pad': 4})

plt.tight_layout()
# plt.imshow()
plt.savefig("sample_chest.png")
```
